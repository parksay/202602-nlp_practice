```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time 
path = "chromedriver.exe"
# service = webdriver.chrome.service.Service(path)
service = webdriver.chrome.service.Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
url = "https://www.youtube.com/watch?v=NtFegtMplyo"


driver.get(url)
time.sleep(5)

# path_container = "div#review-list-area"
path_container = "ytd-comments#comments"
container_review = driver.find_element(By.CSS_SELECTOR, path_container)
path_review = "span[role='text]"
reviews = container_review.find_elements(By.CSS_SELECTOR, path_review)
for item in reviews:
    contents = item
    print(contents)

driver.quit()
```
