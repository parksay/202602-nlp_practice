```python
# 2-2. corpus -> tokenize -> 정규화+cleansing -> 문장
corpus = '영화의 오페라화라고 해야할 작품. 극단적 평갈림은 어쩔 수 없는 듯.'
```

```python
import re
# re.sub(pattern, repl, string, count=0, flags=0)pattern: 치환할 대상(정규식 패턴)
# repl: 바꿀 문자열 또는 함수
# string: 원본 문자열
# count: 치환 횟수 제한 (기본값 0 → 전체 치환)
# flags: 정규식 옵션 (예: re.IGNORECASE)
filtered = re.sub(r"[^0-9a-zA-Z가-힣\s\.]", "", corpus)
filtered
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

# 
# a = [1, 2, 3] 
# b = [4, 5] 
# b.extend(a) # [4, 5, 1, 2, 3]
# list1.extend(list2) 하면 list1 을 넓혀서 list2 까지 포함
# 
sent_list = sent_tokenize(filtered)
word_list = []
for sent in sent_list:
    tokens = word_tokenize(sent)
    word_list.extend(tokens)
word_list
```

```python
import kss
# 
# kss = Korean Sentence Splitter
# kss 는 문장을 자동으로 분리해주는 도구. 근데 이제 한국어에 특화된.
# kss.split_sentences(text) # 문장 단위로 잘라줌
# kss.split_morphemes(text) # 형태소 단위로 잘라줌
sent_list = kss.split_sentences(filtered) # 
sent_list
```

```python
from konlpy.tag import Okt
# konlpy = Korean Natural Language Processing in Python
# Okt = Open Korean Text / 트위터에서 만든 오픈소스 한국어 형태소 분석기
# okt.morphs(filtered) # 형태소 단위로 나눠줌
# okt.pos(filtered) # 형태소 단위로 나눠서 품사를 함께 태깅해줌
# okt.phrases(filtered) # 구 단위로 나눠줌
okt = Okt()
okt.morphs(filtered, stem=True) # 형태소 단위로 나눠주는데 "stem=True" 속성 주면 lemmatization
```
